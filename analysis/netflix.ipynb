El dataset tiene 9827 peliculas, y 9 variables. Las variables son:
- Release_Date: La fecha de estreno de cada pelicula (Fecha)
- Title: Titulo de la pelicula (String)
- Overview: Descripcion de la pelicula (String)
- Popularity: Que tan popular es la pelicula (Number)
- Vote_Count: Cuantas personas votaron que es popular (Int)
- Vote_average: Voto promedio de popularidad (Float)
- Original_Language: Idioma Original (String)
- Genre: Listado de Generos de la pelicula (List of strings)
- Poster_Url: Imagen del poster (string)

```python
#como el poster_url no es importante, lo vamos a droppear

movies_data = movies_data.drop(columns='Poster_Url')
movies_data.head()
```

¿Hay valores faltantes?
¿Hay registros duplicados?

```python
#check for nulls:
print(movies_data.isna().sum())
#no hay valores faltantes
```

```python
#check for duplicates:
duplicates = movies_data[movies_data.duplicated()]
print(duplicates)
#no hay duplicates
```

```python
#desarmar listas de generos:
max_genres = movies_data['Genre'].str.split(', ').str.len().max()

genre_columns = [f'Genre_{i}' for i in range(1, max_genres + 1)]

movies_data[genre_columns] = pd.DataFrame(
    movies_data['Genre'].str.split(', ').tolist(),
    index=movies_data.index
)

movies_data.head(10)
```
